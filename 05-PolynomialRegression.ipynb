```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)

y = 0.5 * x ** 2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y, y_predict)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


def get_pipeline_er(degree, X, y):
    pipeliner = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('lin_reg', LinearRegression())
    ])

    pipeliner.fit(X, y)
    y_predict = pipeliner.predict(X)
    print(mean_squared_error(y, y_predict))

    plt.scatter(X, y)
    plt.plot(np.sort(X[:, 0]), y_predict[np.argsort(X[:, 0])], color='r')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split


def plot_fit(fit_list, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)

    fig, axes = plt.subplots(2, int(len(fit_list) / 2), figsize=(10, 10), sharex=True, sharey=True)
    fig.subplots_adjust(hspace=0.1, wspace=0.1)  # 设置子图的横纵间距
    for i, ax in enumerate(axes.flat):
        pipeliner = Pipeline([
            ('poly', PolynomialFeatures(degree=fit_list[i])),
            ('std_scaler', StandardScaler()),
            ('lin_reg', LinearRegression())
        ])

        pipeliner.fit(X_train, y_train)
        y_train_predict = pipeliner.predict(X_train)
        y_test_predict = pipeliner.predict(X_test)

        ax.scatter(X_train, y_train)
        ax.scatter(X_test, y_test)

        ax.plot(np.sort(X_train[:, 0]), y_train_predict[np.argsort(X_train[:, 0])], color='r')

        mes_train = np.round(mean_squared_error(y_train, y_train_predict), 2)
        r_2_train = np.round(pipeliner.score(X_train, y_train), 2)
        mes_test = np.round(mean_squared_error(y_test, y_test_predict), 2)
        r_2_test = np.round(pipeliner.score(X_test, y_test), 2)
        ax.text(x=0.02, y=0.85,  #文本x、y轴坐标
                s=f"""Train: MSE={mes_train}\t$R^2={r_2_train}$ \nTest: MSE={mes_test}\t$R^2={r_2_test}$""",  #文本内容
                fontdict=dict(fontsize=12),  #字体属性字典
                bbox={  #添加文字背景色
                    'facecolor': 'white',  #填充色
                    'edgecolor': 'gray',  #外框色
                    'alpha': 0.3,  #框透明度
                    'boxstyle': 'round'
                },
                transform=ax.transAxes)

    # 添加图例
    fig.legend(['Training Set', 'Testing Set'], loc='lower center', ncol=2, bbox_to_anchor=(0.5, 0.02))
    plt.show()


plot_fit([1, 2, 10, 20], X, y)
# 可以看到，随着多项式拟合所使用的未知数越多，训练集的拟合精度越来越高，MSE损失越来越小。但是拟合出来的多项式对于测试集数据的泛化能力却逐渐减弱。
```

```python
from sklearn.model_selection import train_test_split


def show_fit(fit_list, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)

    r_2 = []
    for i in range(len(fit_list)):
        pipeliner = Pipeline([
            ('poly', PolynomialFeatures(degree=fit_list[i])),
            ('std_scaler', StandardScaler()),
            ('lin_reg', LinearRegression())
        ])

        pipeliner.fit(X_train, y_train)

        r_2.append((pipeliner.score(X_train, y_train), pipeliner.score(X_test, y_test)))

    return r_2


show_fit([1, 2, 10, 20], X, y)
```

```python
degree = [i for i in range(1, 100)]
fit_list = show_fit(degree, X, y)
plt.plot(degree, [i[0] for i in fit_list])
plt.plot(degree, [i[1] for i in fit_list])
```

## 学习曲线（Learning Curve）

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)

y = 0.5 * x ** 2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
def plot_learning_curve(algo, X_train, X_test, y_train, y_test):
    train_score = []
    test_score = []

    x_plot = [i for i in range(1, len(X_train) + 1)]

    for i in x_plot:
        algo.fit(X_train[:i], y_train[:i])
        y_train_predict = algo.predict(X_train[:i])
        y_test_predict = algo.predict(X_test[:i])
        train_score.append(np.sqrt(mean_squared_error(y_train[:i], y_train_predict)))
        test_score.append(np.sqrt(mean_squared_error(y_test[:i], y_test_predict)))

    plt.plot(x_plot, train_score, label='Train')
    plt.plot(x_plot, test_score, label='Test')
    plt.axis([0, len(X_train) + 1, 0, 4])
    plt.legend()
    plt.show()


plot_learning_curve(LinearRegression(), X_train, X_test, y_train, y_test)
```

## 交叉验证（Cross Validation）

![](https://i.imgur.com/ByoMNP9.png)

![](https://i.imgur.com/epDicRU.png)

```python
from sklearn import datasets

# Load data
digits = datasets.load_digits()
X = digits['data']
y = digits['target']
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
%%time

from sklearn.neighbors import KNeighborsClassifier

best_score = 0
best_k = None
best_p = None
for k in range(2, 11):
    for p in range(1, 6):
        knn_classifier = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p)
        knn_classifier.fit(X_train, y_train)
        score = knn_classifier.score(X_test, y_test)
        if score > best_score:
            best_p = p
            best_k = k
            best_score = score

(best_score, best_k, best_p)
```

```python
from sklearn.model_selection import cross_val_score

knn_classifier = KNeighborsClassifier()
cross_val_score(knn_classifier, X_train, y_train)
```

```python
%%time

from sklearn.neighbors import KNeighborsClassifier

best_score = 0
best_k = None
best_p = None
for k in range(2, 11):
    for p in range(1, 6):
        knn_classifier = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p)
        knn_classifier.fit(X_train, y_train)
        # score = knn_classifier.score(X_test, y_test)
        score = np.mean(cross_val_score(knn_classifier, X_test, y_test))

        if score > best_score:
            best_p = p
            best_k = k
            best_score = score

(best_score, best_k, best_p)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=best_k, p=best_p)
knn_classifier.fit(X_train, y_train)
knn_classifier.score(X_test, y_test)
```

## 偏差方差权衡（Bias Variance Trade off）

![](https://i.imgur.com/XoseML8.png)

```python
# 模型误差 = 偏差(Bias) + 方差(Variance) + 不可避免的误差
# 导致偏差的主要原因:
# 对问题本身的假设不正确！
# 如:非线性数据使用线性回归
# 导致方差的主要原因
# 数据的一点点扰动都会较大地影响模型。
# 通常原因，使用的模型太复杂。
# 如高阶多项式回归。
# 有一些算法天生是高方差的算法。如kNN。
# 非参数学习通常都是高方差算法。因为不对数据进行任何假设
# 有一些算法天生是高偏差算法。如线性回归。
# 参数学习通常都是高偏差算法。因为堆数据具有极强的假设
# 偏差和方差通常是矛盾的。
# 降低偏差，会提高方差。
# 降低方差，会提高偏差。、
# 机器学习的主要挑战，来自于方差！
# 解决高方差的通常手段:
# 1降低模型复杂度
# 2减少数据维度;降噪
# 3 增加样本数
# 4使用验证集
# 5、模型的正则化
```

## 模型正则化 Regularization

模型正则化:限制参数的大小

![](https://i.imgur.com/iUiXVcH.png)

```python
# 上面这种模型正则化的方式被称为——岭回归 Ridge Regression
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)

y = 0.5 * x + 3 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
pipeliner = Pipeline([
    ('poly', PolynomialFeatures(degree=20)),
    ('std_scaler', StandardScaler()),
    ('lin_reg', LinearRegression())
])

pipeliner.fit(X_train, y_train)
y_predict = pipeliner.predict(X_test)
mean_squared_error(y_test, y_predict)
```

```python
def plot_model(model):
    X_plot = np.linspace(-3, 3, 100).reshape(100, 1)
    y_plot = model.predict(X_plot)

    plt.scatter(x, y)
    plt.plot(X_plot[:, 0], y_plot, c='r')
    plt.axis([-3, 3, 0, 6])
    plt.show()

plot_model(pipeliner)
```

```python
# 使用岭回归
from sklearn.linear_model import Ridge

def ridge_regression(degree, alpha):
    return Pipeline([
    ('poly', PolynomialFeatures(degree=20)),
    ('std_scaler', StandardScaler()),
    ('lin_reg', Ridge(alpha=alpha))
])

ridge_regression_er = ridge_regression(20, 1)
ridge_regression_er.fit(X_train, y_train)

y_predict = ridge_regression_er.predict(X_test)
mean_squared_error(y_test, y_predict)
```

```python
plot_model(ridge_regression_er)
```

```python
ridge_regression_er = ridge_regression(20, 100000)
ridge_regression_er.fit(X_train, y_train)

plot_model(ridge_regression_er)  # 当alpha无穷大的时候，对于优化函数其实就是只有theta全为0，才可以保证优化到最小值
```
